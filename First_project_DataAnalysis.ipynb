```python
import pandas as pd 
import numpy as np 
import plotly.express as px 
from matplotlib import pyplot as plt 
import seaborn as sns
pd.options.display.float_format = '{:,.2f}'.format 
pd.options.display.max_rows = None 
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('all_data.csv', encoding= 'iso-8859-1')
```

```python
df.info()
```

```python
df.sample(10)
```

```python
# الاخطاء الحالية  
# Delete orders with NaN values
#Rename City.1, Country.1, Phon.1 Columns
#Change OrderDate Type into Date
#Change ID Columns, and OrderNumber column to integer
```

```python
#نشوف الخانات الفاضية في order id ونمسحها 
df[df.OrderID.isnull()]
```

```python
# Delete Orders NaN Rows
df.dropna(subset='OrderID', inplace=True)
```

```python
df.info()
```

```python
# Rename Columns
df.rename(columns={ 'City.1' : 'SuppCity', 'Country.1' : 'SuppCountry', 'Phone.1' : 'SuppPhone'}, inplace= True)
```

```python
# Change OrderDate and ID Columns type to Date, integer
df.OrderDate = pd.to_datetime(df.OrderDate)   
```

```python
df.columns
```

```python
#هنحول الارقام من float to integer
for columns in df.columns:
    if 'ID' in columns or 'OrderNumber' in columns:
        df[columns] = df[columns].astype(int)
```
